```python
import os
import numpy as np
import cv2


path_images = './Helen_testset/y_pred/'
path_pts = './Helen_testset/y_true/'
images = os.listdir(path_images)
images.sort()
pts = os.listdir(path_pts)
pts.sort()
print(len(images), len(pts))


def read_pts(filename):
  path_pts = './Helen_testset/y_true/'
  return np.loadtxt(path_pts + filename, comments=("version:", "n_points:", "{", "}"))
#print(obtener_landmarksMediapipe(images[0]))  # Obtenemos los landmarks

def errorpercent(prediction, truth):
  sum = 0
  for i in range(len(truth)):
    xp = (100/truth[i,0]) * abs(truth[i,0]-prediction[i,0])
    yp = (100/truth[i,1]) * abs(truth[i,1]-prediction[i,1])
    sum = sum + xp + yp
  prom = sum/(len(truth)*2)
  return prom

def NME(prediction, truth):
  error = 0
  sum = 0
  for i in range(len(truth)):
    xp = (1/truth[i,0]) * abs(truth[i,0]-prediction[i,0])
    yp = (1/truth[i,1]) * abs(truth[i,1]-prediction[i,1])
    sum = sum + xp + yp
  prom = sum/(len(truth)*2)
  if prom > 0.05:
    error=1
  return error

print("End ------------- End")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from models import mediapipe_model
from utils import find_face

df = pd.DataFrame()
df = pd.DataFrame(columns=['NUMERO', 'IMG', 'PTS', 'NME', 'ERROR'])
for i in range(0,5):
    path_img = path_images + images[i]
    #imag1 = cv2.imread(path_img)
    var_truth = read_pts(pts[i])
    imag,  truth = find_face.my_face_found_GrandTruth(var_truth,path_img, 100)
    truth = np.array(truth)
    y_pred = []
    landmarks, Df = mediapipe_model.MediapipeModel(var_truth, path_img, y_pred)
    error = str(errorpercent(landmarks, truth))
    nme = str(NME(landmarks,truth))
    print("Imagen; "+path_img+" pts: "+pts[i]+" Numero: "+str(i))
    print("% Error: "+error)
    print("NME: "+nme)
    df = df.append({'NUMERO': i, 'IMG': path_img, 'PTS':pts[i], 'NME': nme, 'ERROR': error}, ignore_index=True)
    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(truth[:,0],truth[:,1], marker='o',s=10, color="b", label="Grand truth\n%Error: "+error+"\nNME: "+nme)
    ax.imshow(imag)
    ax.scatter(landmarks[:,0],landmarks[:,1], marker='o',s=5, color="r", label="Prediction: "+str(i))
    ax.legend()
    # Resultado falla la imagen 82
```
